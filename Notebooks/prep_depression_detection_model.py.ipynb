```python
#normalization

def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

#train_data_nan_handled3 = normalize(train_data_nan_handled2)
#print type(train_data_nan_handled2['profile_imageFeatures.BlueChannelMean'].max())
#print train_data_nan_handled3.head()

#selecting only numerical values
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
train_data_nan_handled3 = train_data_nan_handled2.select_dtypes(include=numerics)
#print train_data_nan_handled3.head()
train_data_nan_handled4 = normalize(train_data_nan_handled3)
print train_data_nan_handled4.shape
print train_data_nan_handled4['profile_imageFeatures.colorfulness'].head()
```
